```python
import pandas as pd
```

```python
edges_df = pd.read_csv("user_edges.csv")
edges_df.head()
```

```python
nodes_df = pd.read_csv("user_nodes_new.csv")
nodes_df.head()
```

```python
id_label_dict = {}
for row in nodes_df.itertuples():
    id_label_dict[row.Id] = row.Label
id_label_dict
```

```python
source = []
target = []
label = []

for row in edges_df.itertuples():
    source.append(id_label_dict[row.Source])
    target.append(id_label_dict[row.Target])
    label.append(row.Label)
```

```python
# create the dataframe
tweets_df = pd.DataFrame({'Source': source, 'Target': target, 'Label': label})

# passe to csv
tweets_df.to_csv("tweets.csv", index=False)
```
