```python
import pandas as pd
```

```python
user_analysis_df = pd.read_csv("user_analysis.csv")
user_analysis_df
```

```python
sentiment_traduction_dict  = {'sentiment': {'Negative':-1, 'Neutral':0, 'Positive':1}}
user_analysis_df = user_analysis_df.replace(sentiment_traduction_dict)
user_analysis_df
```

```python
classified_users_df = user_analysis_df.groupby('user_name')[['class_with', 'class_without', 'sentiment']].median()
classified_users_df['class_with'] = classified_users_df['class_with'].astype(int)
classified_users_df['class_without'] = classified_users_df['class_without'].astype(int)
classified_users_df['sentiment'] = classified_users_df['sentiment'].astype(int)
classified_users_df
```

```python
classes_sentiment_df = classified_users_df.groupby('class_without')[['sentiment']].mean()
classes_sentiment_df
```

```python
classes_sentiment_df = classified_users_df.groupby('class_with')[['sentiment']].mean()
classes_sentiment_df[-1:]
```

```python
# passe to csv
classified_users_df.to_csv("classified-users.csv")
```
