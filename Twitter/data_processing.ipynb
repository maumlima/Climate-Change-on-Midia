```python
# create nodes
users=[] # users' id, name, description, character(occupation/institutionÔºâ
tweets = [] # tweets' id, text
deja=[]

for index,row in data.iterrows():
    if row['user_id'] not in deja:
        users.append([row['user_id'],row['user_name'],row['user_description'],' '])
        deja.append(row['user_id'])
    tweets.append([row['tweet_id'],row['content']])

#df = pd.DataFrame(tweets, columns=['Id','Label']) # label means the content of the tweet
#df.to_csv('tweet_nodes.csv',index=False)
```

```python
zidian =  dict([(i,a) for i,a in zip(data['tweet_id'], data['user_id'])])
```

```python
# complete the nodes
for index,row in data.iterrows():

    if not row['quotedTweet_id']==0:#this tweet quoated someone
        if row['quotedTweet_id'] in zidian.keys():
            continue
        else:
            for t in snst.TwitterTweetScraper(row['quotedTweet_id']).get_items():
                u = t.user
                zidian[row['quotedTweet_id']]=u.id
            if not u.id in deja:
                users.append([u.id,u.username,u.description,' ']) 
                deja.append(u.id)
            
   
    if not type(row['mentionedUsers'])is float:
        m_users = eval(row['mentionedUsers'])
        for i in range(len(m_users)):
            if not m_users[i].id in deja:
                users.append([m_users[i].id,m_users[i].username,m_users[i].description,' ']) 
                deja.append(m_users[i].id)
```

```python
df = pd.DataFrame(users, columns=['Id','Label','Description','Character']) # label means the name of the user
df.to_csv('user_nodes.csv',index=False)
```

```python
# create relationships between users
edges=[]
for index,row in data.iterrows():
    # relation quoted
    if not row['quotedTweet_id']==0:
        if row['quotedTweet_id'] in zidian.keys():
            u_id= zidian[row['quotedTweet_id']]
            edge = [row['user_id'],u_id,'quoted']
            edges.append(edge)
    # relation mentioned
    if not type(row['mentionedUsers'])is float:
        m_users = eval(row['mentionedUsers'])
        for i in range(len(m_users)):
            edge = [row['user_id'],m_users[i].id,'mentioned']
            edges.append(edge)
user_edges = pd.DataFrame(edges, columns=['Source','Target','Label'])
user_edges.to_csv('user_edges.csv',index=False)
```

```python
# collect the contents
zidian2 = {}
for index,row in data.iterrows(): 
    if row['user_id'] in zidian2.keys():
        zidian2[row['user_id']].append(row['content'])
    else:
        zidian2[row['user_id']]=[row['content']]
```

```python
my_list=[]
for key in zidian2.keys():
    my_list.append([key,zidian2[key]])
my_df = pd.DataFrame(my_list, columns=['user_id','content'])
```

```python
my_df
```

```python
# text cleaning
# making all tweets lowercase
def lowercase_word(text):
    text  = "".join([char.lower() for char in text])
    return text

for index,row in my_df.iterrows():
    my_df.loc[index,'content']= lowercase_word(row['content'])

#removing hyperlinks mentioned in the tweets
def remove_links(text):
    processed_text = re.sub(r"(?:\@|http?\://|https?\://|www)\S+", "", text)
    processed_text = " ".join(processed_text.split())
    return processed_text

for index,row in my_df.iterrows():
       my_df.loc[index,'content']= remove_links(row['content'])

#removing tagged users from the tweets
def remove_users(text):
    processed_text = re.sub(r'@\w+ ?'," ",text)
    processed_text = " ".join(processed_text.split())
    return processed_text

for index,row in my_df.iterrows():
    my_df.loc[index,'content']= remove_users(row['content'])

#removing hashtags related to climate change
def rem_hashtags(text):
    processed_text = re.sub(r"#climatechange", " ", text)
    #processed_text = re.sub(r"amp", " ", text)
    processed_text = " ".join(processed_text.split())
    return processed_text

for index,row in my_df.iterrows():
    my_df.loc[index,'content']= rem_hashtags(row['content'])
    

#removing punctuations in the tweets
def remove_punct(text):
    punctuations = '!"#$%&\'()*+,-./:;<=>?@[\\]^_`{|}~'
    text  = "".join([char for char in text if char not in punctuations])
    text = re.sub('[0-9]+', '', text)
    return text
for index,row in my_df.iterrows():
    my_df.loc[index,'content']= remove_punct(row['content'])

# removing emoji by words
import emoji
for index,row in my_df.iterrows():
    tweet = row['content']
    tweet= emoji.demojize(tweet)
    tweet = tweet.replace(":"," ")
    tweet = ' '.join(tweet.split())
    my_df.loc[index,'content']= tweet
```
